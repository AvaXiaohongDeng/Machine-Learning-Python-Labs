```python
#loading data from a URL
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
# Define the column names
col_names = ["sepal_length_in_cm", "sepal_width_in_cm",  "petal_length_in_cm",  "petal_width_in_cm",  "class"]
# Read data from URL
iris_data = pd.read_csv(url, names=col_names)
# print first five rows from iris_data 
iris_data.head(10)
```

```python
# generates descriptive statistics, including those that summarize the central tendency, 
# dispersion, and shape of the dataset’s distribution
iris_data.describe()
```

```python
#use the pandas .apply() method. This function acts as a map() function in Python. 
#It takes a function as an input and applies this function to an entire DataFrame.
df = pd.DataFrame({'A': [1, 2], 'B': [10, 20]})

def square(x):
    return x * x

df1 = df.apply(square)
print(df)
print(df1)
```

```python
#Using a lambda function. We can create a lambda function while calling the apply() function.
df2 = df.apply(lambda x: x * x)
df2
```

```python
#simple aggregation
import pandas as pd
washers = pd.read_csv("washers.csv")
washers[['BrandName']].describe()                        #get simple aggregation for a column in a dataframe
washers[['BrandName']].value_counts(normalize = True)  #count the brands but normalize It
washers[['Volume']].mean()                             #get the averae of a column
washers.groupby('BrandName')[['Volume']].mean()        #get group level aggregation

```

**Task 4: Data Normalization**

4.1 Explain why normalization is important in machine learning.

Normalization is crucial in machine learning for the following reasons:

* Some models achieve better performance when features are on a similar scale, facilitating faster algorithm convergence.
* Normalization ensures that all features contribute equally to the model, preventing dominance by those with larger magnitudes.
* Normalization addresses sensitivity issues in algorithms like k-nearest neighbors and support vector machines, which are influenced by the scale of input features.

**Task 5: Addressing Class Imbalance**

5.1 Discuss why class imbalance can be a problem in machine learning models.

Class imbalance can be a problem in machine learning models because standard learners often exhibit bias towards the majority class. They prioritize minimizing global quantities like error rates without considering the underlying data distribution. Consequently, examples from the majority class are well-classified, while those from the minority class tend to be misclassified.
<br>

5.2 Use a technique to address class imbalance (e.g., oversampling, undersampling).
1. Load the prepared dataset

2. Identify if there is class imbalance issue via 
    - Calculating the distribution of each class of the target column.
    - Displaying the discribution by pie chart.

3. Choose a technique to address class imbalance.

```python
#Collect data by loading a CSV file from the current working directory
class_unbalanced_iris = pd.read_csv("Class_unbalanced_iris_data.csv")
class_unbalanced_iris.head(2)
```

```python
#calculate the distribution of each class of the target column
class_unbalanced_iris['class'].value_counts() 
```

```python
#display pie chart
class_unbalanced_iris['class'].value_counts(normalize=True).plot.pie(autopct='%.2f')
```

**<span style="color:blue;">Observations:</span>**

We can see that the target column("class") has class imbalance issue. The percentage of the majority class("Iris-versicolor") is clearly different from the other two classes. 

Solution:  
The 'undersampling' technique will be employed to simplify, improve readability, and expedite the resolution of the class imbalance.

5.3 Implement the technique using an appropriate Python package.

```python
from imblearn.under_sampling import RandomUnderSampler

#Create an undersampler
undersampler = RandomUnderSampler(sampling_strategy='auto', random_state=42)

#Create features (X) and target variable (y)
X = class_unbalanced_iris.drop(columns=['class'])
y = class_unbalanced_iris['class']

#Employ undersampling
X_resampled, y_resampled = undersampler.fit_resample(X, y)

#Create a new DataFrame with the resampled data
iris_resampled = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), pd.Series(y_resampled, name='class')], axis=1)

#Displaying the discribution after undersampling by pie chart.
iris_resampled['class'].value_counts()
```

**<span style="color:blue;">Observations:</span>**

We can see that some instances of "Iris-versicolor" and "Iris-virginice" have been randomly removed so that all three classes have the same size.
